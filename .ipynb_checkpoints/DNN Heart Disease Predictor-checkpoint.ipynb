# Preprocess our dataset

##### We have to normalize only some continuous feature columns and to employ feature engineering to make the dataset simple.

```python
df.describe()
```

```python
df = (df - df.min())/(df.max() - df.min())
```

```python
df.head()
```

##### Now, merging the categorical columns into one 

# Split our dataset

##### Now, let's split our dataset for training and evaluation using train_test_split from scikit learn

```python
from sklearn.model_selection import train_test_split
```

##### We want to separate our features and labels to use this method 

```python
df = df.sample(frac= 1).reset_index(drop= True)
```

```python
labels = df['target']

features = df.drop(columns= 'target')
```

```python
labels.shape
```

```python
features.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, labels, test_size= 0.2, shuffle= True)
```

```python
print ("x_train {}".format(x_train.shape))
print ("y_train {}".format(y_train.shape))
print ("x_test {}".format(x_test.shape))
print ("y_test {}".format(y_test.shape))
```

# Creating feature columns for our features

##### Feature columns are created so that the data flow in tensorflow methods will be smooth and flexible. It has many other features like bucketizing the data, one hot encode our data, etc.

```python
feature_columns = [tf.feature_column.numeric_column(x) for x in features.columns]
```

```python
print (feature_columns)
```

# Creating input functions for training and evaluating

```python
Train_input_fn = tf.estimator.inputs.pandas_input_fn(x= x_train,
                                               y = y_train,
                                               batch_size= 27,
                                               num_epochs= 100,
                                               shuffle= True)
eval_input_fn = tf.estimator.inputs.pandas_input_fn(x = x_test,
                                                    y= y_test,
                                                    num_epochs=1,
                                                    batch_size=10,
                                                    shuffle= False)
```

# Instantiating DNN Classifier with the feature columns 

```python
model_dir = './Trained'
model = tf.estimator.DNNClassifier(hidden_units=[100, 75, 50, 25, 12, 2],
                                feature_columns= feature_columns,
                                model_dir= model_dir,
                                n_classes= 2,
                                dropout= 0.2)
```

```python
shutil.rmtree(model_dir, ignore_errors= True)

model.train(train_input_fn)
```

```python
model.evaluate(eval_input_fn)
```
